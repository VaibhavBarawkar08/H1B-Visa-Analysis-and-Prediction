```python
df['FULL_TIME_POSITION'] = df['FULL_TIME_POSITION'].map({'Y': 1, 'N': 0})
df['H1B_DEPENDENT']=df['H1B_DEPENDENT'].map({'Yes':1,'No':0})
df['WILLFUL_VIOLATOR']=df['WILLFUL_VIOLATOR'].map({'Yes':1,'No':0})
# Convert multi-value features to binary
df['NEW_EMPLOYMENT'] = df['NEW_EMPLOYMENT'].apply(lambda x: 1 if x > 0 else 0)
df['CONTINUED_EMPLOYMENT'] = df['CONTINUED_EMPLOYMENT'].apply(lambda x: 1 if x > 0 else 0)
df['CHANGE_PREVIOUS_EMPLOYMENT'] = df['CHANGE_PREVIOUS_EMPLOYMENT'].apply(lambda x: 1 if x > 0 else 0)

```

```python
X=df[selected_features]
y=df['CASE_STATUS_BINARY']
```

```python
del df
import gc
gc.collect()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=100)
```

```python
del X,y
gc.collect()
```

```python
print("X_train:",X_train.shape)
print("y_train:",y_train.shape)
print("X_test:",X_test.shape)
print("y_test:",y_test.shape)
```

```python
from sklearn.utils import resample
import pandas as pd

# Combine X_train and y_train for easy handling
train_df = pd.concat([X_train, y_train], axis=1)

# Separate classes
majority = train_df[train_df['CASE_STATUS_BINARY'] == 1]
minority = train_df[train_df['CASE_STATUS_BINARY'] == 0]

# Downsample majority class
majority_downsampled = resample(
    majority,
    replace=False,
    n_samples=len(minority),
    random_state=42
)

# Combine downsampled majority with minority
balanced_train_df = pd.concat([majority_downsampled, minority])

# Shuffle the new training data
balanced_train_df = balanced_train_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Separate X and y again
X_train_down = balanced_train_df.drop('CASE_STATUS_BINARY', axis=1)
y_train_down = balanced_train_df['CASE_STATUS_BINARY']
```

```python
print("Original class distribution:")
print(y_train.value_counts())

print("\nAfter downsampling:")
print(y_train_down.value_counts())
```

```python
del balanced_train_df
gc.collect()
```

```python
# Step 4: Define transformers
categorical_features = ['JOB_TITLE']
numeric_features = ['ANNUAL_WAGE', 'EMPLOYMENT_DURATION_DAYS']
ordinal_features = ['PW_WAGE_LEVEL']
binary_features = [
    'FULL_TIME_POSITION',
    'NEW_EMPLOYMENT',
    'CONTINUED_EMPLOYMENT',
    'CHANGE_PREVIOUS_EMPLOYMENT',
    'CHANGE_EMPLOYER',
    'NEW_CONCURRENT_EMPLOYMENT',
    'AMENDED_PETITION'
]
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Pipelines
pw_wage_order = [['I', 'II', 'III', 'IV', 'V']] 

# Pipelines
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

ordinal_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal', OrdinalEncoder(categories=pw_wage_order))
])

binary_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent'))
])

# Combine into column transformer
preprocessor = ColumnTransformer(transformers=[
    ('cat', categorical_transformer, categorical_features),
    ('num', numeric_transformer, numeric_features),
    ('ord', ordinal_transformer, ordinal_features),
    ('bin', binary_transformer, binary_features)
])

```

```python
X_train_processed = preprocessor.fit_transform(X_train_down)
X_test_processed = preprocessor.transform(X_test)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
```

```python
# Define classifiers
classifiers = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "AdaBoost": AdaBoostClassifier(random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    "Naive Bayes": GaussianNB()
}

# Build pipelines for all classifiers
pipelines = {
    name: Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', clf)
    ]) for name, clf in classifiers.items()
}
```

```python
for name, pipeline in pipelines.items():
    print(f"Training {name}...")
    pipeline.fit(X_train_down, y_train_down)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

for name, pipeline in pipelines.items():
    print(f"\n {name}")
    y_pred = pipeline.predict(X_test)
    y_proba = pipeline.predict_proba(X_test)[:, 1] if hasattr(pipeline.named_steps['classifier'], "predict_proba") else None
    
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    
    if y_proba is not None:
        print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
```

```python
from joblib import dump

for name, pipeline in pipelines.items():
    dump(pipeline, f"{name.lower().replace(' ', '_')}_pipeline.joblib")
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, roc_auc_score
```

```python
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

pipeline = Pipeline([
    ('preprocessing', preprocessor),
    ('classifier', xgb)
])
```

```python
param_dist = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [3, 5, 7],
    'classifier__learning_rate': [0.01, 0.05, 0.1],
    'classifier__subsample': [0.7, 0.8, 1.0],
    'classifier__colsample_bytree': [0.6, 0.8, 1.0],
    'classifier__scale_pos_weight': [1, 2, 3]  # Helpful for class imbalance
}
```

```python
search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_dist,
    n_iter=20,             # You can increase this if time permits
    scoring='roc_auc',
    cv=3,
    verbose=2,
    n_jobs=-1,
    random_state=42
)

search.fit(X_train_down, y_train_down)
```

```python
best_model = search.best_estimator_

# Transform test set
X_test_transformed = preprocessor.transform(X_test)

# Predict
y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:, 1]

# Evaluate
print("Classification Report:\n", classification_report(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
```
