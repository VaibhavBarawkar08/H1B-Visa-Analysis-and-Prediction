```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("cleaned_and_preprocessed.csv")
```

```python
df['CASE_STATUS_BINARY'].value_counts().plot(kind='bar')
```

```python

df['WORKSITE_CITY'].value_counts().head(10).plot(kind='barh')
```

```python
df['EMPLOYER_NAME'].value_counts().head(10).plot(kind='barh')
```

```python
df['JOB_TITLE'].value_counts().head(10).plot(kind='barh')
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='FULL_TIME_POSITION', hue='CASE_STATUS_BINARY')
plt.title('Visa Case Status by Full-Time Position')
plt.xlabel('Full Time Position')
plt.ylabel('Number of Applications')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='H1B_DEPENDENT', hue='CASE_STATUS_BINARY')
plt.title('Visa Case Status by H1B Dependent')
plt.xlabel('H1B Dependent')
plt.ylabel('Number of Applications')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='WILLFUL_VIOLATOR', hue='CASE_STATUS_BINARY')
plt.title('Visa Case Status by Willful Violator')
plt.xlabel('Willful Violator')
plt.ylabel('Number of Applications')
plt.tight_layout()
plt.show()
```

```python


pw_level_df = pd.crosstab(df['PW_WAGE_LEVEL'], df['CASE_STATUS_BINARY'])
pw_level_df.plot(kind='bar', stacked=True, figsize=(8, 5))
plt.title('Visa Case Status by Wage Level')
plt.xlabel('Wage Level')
plt.ylabel('Number of Applications')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
worker_bin_df = pd.crosstab(df['WORKER_COUNT_BIN'], df['CASE_STATUS_BINARY'])
worker_bin_df.plot(kind='bar', stacked=True, figsize=(8, 5))
plt.title('Visa Case Status by Worker Count Bins')
plt.xlabel('Worker Count Bin')
plt.ylabel('Number of Applications')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_soc_titles = df['SOC_TITLE'].value_counts().nlargest(10).index
filtered_df = df[df['SOC_TITLE'].isin(top_soc_titles)]

plt.figure(figsize=(10, 6))
sns.countplot(data=filtered_df, y='SOC_TITLE', hue='CASE_STATUS_BINARY', order=top_soc_titles)
plt.title('Visa Case Status by Top 10 SOC Titles')
plt.xlabel('Number of Applications')
plt.ylabel('SOC Title')
plt.legend(title='Case Status')
plt.tight_layout()
plt.show()
```
